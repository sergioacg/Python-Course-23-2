```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])
```

Crear Datos y Graficarlos

```python
x = np.arange(0, 10, 0.1)
y = 2 * np.sin(x)
plt.plot(x, y, label = '2sin(x)')
plt.legend()
plt.title('Señal Frecuencial', fontsize= 16)
plt.xlabel('Tiempo (s)', fontsize= 14)
plt.ylabel('Amplitud', fontsize= 14)
plt.rcParams.update({'font.size': 18}) #Modifica el tamaño de todo el notebook 
plt.grid()
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y = 2 * np.sin(x) #Seno
z = np.cos(4*x) #Coseno
plt.plot(x, y, x, z)
plt.legend(['2sin(x)', 'cos(4x)'], loc = 1)
plt.title('Señal Frecuencial', fontsize= 16)
plt.xlabel('Tiempo (s)', fontsize= 14)
plt.ylabel('Amplitud', fontsize= 14)
plt.rcParams.update({'font.size': 18}) #Modifica el tamaño de todo el notebook 
plt.grid()
plt.show()
```

```python
plt.plot(x, y)
plt.title('Seno')
plt.show()
plt.figure()
plt.plot(x, z)
plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot(x,y)
plt.subplot(212)
plt.plot(x, z)
plt.show()
```

```python
plt.plot(x, y, '-m', linewidth=10)
plt.plot(x, z, ':c', linewidth=6)
plt.show()
```

```python
fig, ax = plt.subplots()
plt.plot(x, y, '-m', linewidth=3)
plt.plot(x, z, '--k', linewidth=3)
ax.text(4, 0.5, 'Funciones en Frecuencia', fontsize=12)
ax.axis([3, 7, -1, 1])
plt.show()
```

```python
plt.plot(x, y, '-r', linewidth=3)
plt.title('Función y = 2 sin($\\omega$t)')
plt.show()
```
